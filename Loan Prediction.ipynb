```python
married=pd.get_dummies(x['Married'],drop_first=True)
gender=pd.get_dummies(x['Gender'],drop_first=True)
educ=pd.get_dummies(x['Education'],drop_first=True)
Self=pd.get_dummies(x['Self_Employed'],drop_first=True)
Property=pd.get_dummies(x['Property_Area'],drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
log_model=LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
log_pred=log_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(accuracy_score(y_test,log_pred))
print(classification_report(y_test,log_pred))
print(confusion_matrix(y_test,log_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model=DecisionTreeClassifier()
tree_model.fit(x_train,y_train)
```

```python
tree_pred=tree_model.predict(x_test)
```

```python
print(accuracy_score(y_test,tree_pred))
print(classification_report(y_test,tree_pred))
print(confusion_matrix(y_test,tree_pred))
```

```python
print(accuracy_score(y_test,knn_pred))
print(classification_report(y_test,knn_pred))
print(confusion_matrix(y_test,knn_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
random=RandomForestClassifier()
random.fit(x_train,y_train)
rand_pred=random.predict(x_test)
```

```python
print(accuracy_score(y_test,knn_pred))
print(classification_report(y_test,knn_pred))
print(confusion_matrix(y_test,knn_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.model_selection import KFold
```

```python
cv=KFold(n_splits=5,random_state=None,shuffle=False)
```

```python
n_estimators=[int(x) for x in np.linspace(start=300,stop=2000,num=10)]
# Number of features to consider
max_features=['auto','sqrt','log2']
# maximum number of levels in tree
max_depth=[int(x) for x in np.linspace(10,1000,num=10)]
# Minimum number of samples required to split
min_samples_split=[1,3,5,7,9,10,12,15]
# minimum number samples required at each node
min_samples_leaf=[1,3,5,7,0.1,0.5,0.2,10,0.1]


random_grid={'n_estimators':n_estimators,
            'max_features':max_features,
            'max_depth':max_depth,
            'min_samples_split':min_samples_split,
            'min_samples_leaf':min_samples_leaf,
            'criterion':['entrophy','gini']}
```

```python
random_grid
```

```python
rand=RandomizedSearchCV(random,param_distributions=random_grid,cv=cv,scoring='roc_auc',n_iter=50,verbose=2,n_jobs=-1)
```

```python
rand.fit(x_train,y_train)
```

```python
rand.best_estimator_
```

```python
rand.best_params_
```

```python
rand_best=rand.best_estimator_
pred_rand=rand_best.predict(x_test)
```

```python
print(accuracy_score(y_test,pred_rand))
print(classification_report(y_test,pred_rand))
print(confusion_matrix(y_test,pred_rand))
```

```python
params={'criterion':[rand.best_params_['criterion']],
       'max_depth':[rand.best_params_['max_depth']],
       'max_features':[rand.best_params_['max_features']],
       'min_samples_split':[rand.best_params_['min_samples_split']-1,
                           rand.best_params_['min_samples_split'],
                           rand.best_params_['min_samples_split']+1,
                           rand.best_params_['min_samples_split']+2],
        'min_samples_leaf':[rand.best_params_['min_samples_leaf']-2,
                            rand.best_params_['min_samples_split']-1,
                            rand.best_params_['min_samples_leaf'],
                            rand.best_params_['min_samples_leaf']+1,
                            rand.best_params_['min_samples_leaf']+2],
        'n_estimators':[rand.best_params_['n_estimators']-200,
                       rand.best_params_['n_estimators']-100,
                       rand.best_params_['n_estimators'],
                       rand.best_params_['n_estimators']+100,
                       rand.best_params_['n_estimators']+200]
       }

print(params)
```

```python
grid_model=GridSearchCV(random,param_grid=params,cv=cv,n_jobs=-1,verbose=3)
```

```python
grid_model.fit(x_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_estimator_
```

```python
grid_best=grid_model.best_estimator_
```

```python
grid_pred=grid_best.predict(x_test)
```

```python
print(accuracy_score(y_test,grid_pred))
print(classification_report(y_test,grid_pred))
print(confusion_matrix(y_test,grid_pred))
```
